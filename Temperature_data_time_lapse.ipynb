```python
from mpl_toolkits.basemap import Basemap
from netCDF4 import Dataset
import matplotlib.pyplot as plt
import PIL
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = Dataset('/content/APHRO_MA_TAVE_025deg_V1808.1962.nc')
data
```

```python
lons = data.variables['lon'][:]
```

```python
lats = data.variables['lat'][:]
```

```python
time = data.variables['time'][:]
```

```python
tave = data.variables['tave'][:]
```

```python
mp = Basemap(projection='merc', llcrnrlon= 55.716422, llcrnrlat= 6.661470 , urcrnrlon=101.352372 , urcrnrlat= 36.157385 ,resolution= 'h')
```

```python
lon,lat = np.meshgrid(lons,lats)
x,y = mp(lon,lat)
```

```python
days = np.arange(0,365)

for i in days:

  c_scheme = mp.pcolor(x,y, np.squeeze(tave[i,:,:]), cmap = 'jet')
  mp.drawcoastlines()
  mp.drawstates()
  mp.drawcountries()

  cbar = mp.colorbar(c_scheme, location= 'right', pad= '10%')
  day = i+1

  plt.title('Average Temperature: Day' + str(day) + 'of year 1962')
  plt.clim(-40,40)
  plt.savefig(str(day) + '.jpg')
  plt.clf()
```

```python
image_frames = []  #empty list
days = np.arange(1,366)
for k in days:
    new_frame = PIL.Image.open('/content/' +  str(k) +  '.jpg')
    image_frames.append(new_frame)
image_frames[0].save('temperature_timelapse.gif', format = 'GIF', append_images = image_frames[1: ], save_all= True, duration=100, loop=0)
```
